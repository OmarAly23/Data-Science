# U.S. Medical Insurance Costs

```python
import matplotlib.pyplot as plt
```

Look over the dataset

```python
import csv

list_of_insurance = []
with open('insurance.csv') as insurance:
    insurance_reader = csv.DictReader(insurance)
    for row in insurance_reader:
        print(row)
        list_of_insurance.append(row)
```

Analyze the amount of smokers with children regardless of sex that are smokers



```python
num_of_smokers_with_children = 0
male_smokers_with_children = 0
female_smokers_with_children = 0
for element in list_of_insurance:
    if element['smoker'] == 'yes' and element['children'] != '0':
        num_of_smokers_with_children += 1
        if element['sex'] == 'male':
            male_smokers_with_children += 1
        if element['sex'] == 'female':
            female_smokers_with_children += 1

print(f'Total number of smokers with children = {num_of_smokers_with_children}')
print(f'Male Smokers with children = {male_smokers_with_children}')
print(f'Female Smokers with children = {female_smokers_with_children}')

print(f'Percentage of men: {round((male_smokers_with_children/num_of_smokers_with_children)*100, 2)} %')
print(f'Percentage of men: {round((female_smokers_with_children/num_of_smokers_with_children)*100, 2)} %')




labels = ['male smokers with children', 'female smokers with children']
  
# portion covered by each label
slices = [male_smokers_with_children, female_smokers_with_children]
  
# color for each label
colors = ['r', 'b']
  
# plotting the pie chart
plt.pie(slices, labels = labels, colors=colors, 
        startangle=90, shadow = True, explode = (0, 0),
        radius = 1.2, autopct = '%1.1f%%')
  
# plotting legend
plt.legend()
  
# showing the plot
plt.show()
```

Rate of male smokers vs female smokers

```python
total_of_male_smokers = 0
total_of_female_smokers = 0
total = len(list_of_insurance)
for element in list_of_insurance:
    if element['sex'] == 'male' and element['smoker'] == 'yes':
        total_of_male_smokers += 1
    if element['sex'] == 'female' and element['smoker'] == 'yes':
        total_of_female_smokers += 1
        
        
print(f'Total of male smokers: {total_of_male_smokers}')
print(f'Total of female smokers: {total_of_female_smokers}')
print(f'Male smokers make up {round((total_of_male_smokers/total)*100, 2)} % of the population')
print(f'Female smokers make up {round((total_of_female_smokers/total)*100, 2)} % of the population')
print(f'Entire Dataset: {total}')


labels = ['male smokers', 'female smokers']
  
# portion covered by each label
slices = [total_of_male_smokers, total_of_female_smokers]
  
# color for each label
colors = ['r', 'b']
  
# plotting the pie chart
plt.pie(slices, labels = labels, colors=colors, 
        startangle=90, shadow = True, explode = (0, 0),
        radius = 1.2, autopct = '%1.1f%%')
  
# plotting legend
plt.legend()
  
# showing the plot
plt.show()
```
